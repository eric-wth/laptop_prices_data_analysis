# **3. Data Cleaning**

```python
from scipy import stats

# Calculate Z-scores of each value in the DataFrame (for numerical columns)
z_scores = stats.zscore(df.select_dtypes(include=['float64', 'int64']))

# Identify outliers (Z-score > 3 or < -3)
df_outliers = df[(z_scores > 3).any(axis=1) | (z_scores < -3).any(axis=1)]
print(f'Outliers detected: {len(df_outliers)}')
```

```python
# Remove rows with outliers based on Z-score
df_no_outliers = df[(z_scores < 3).all(axis=1) & (z_scores > -3).all(axis=1)]
```

# **4. Analysis**

**What is the distribution of laptop prices?**

```python
import warnings
warnings.filterwarnings('ignore')

# Plot the distribution of prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Price (Euro)'], kde=True)
plt.title('Distribution of Laptop Prices')
plt.xlabel('Price (Euro)')
plt.ylabel('Frequency')
plt.show()
```

**Is there a relationship between RAM and laptop price?**

```python
# Scatter plot of RAM vs Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='RAM (GB)', y='Price (Euro)', data=df)
plt.title('RAM vs Laptop Price')
plt.xlabel('RAM (GB)')
plt.ylabel('Price (Euro)')
plt.show()
```

**Which companies have higher laptop prices?**

```python
# Boxplot of Company vs Price
plt.figure(figsize=(14, 7))
sns.boxplot(x='Company', y='Price (Euro)', data=df)
plt.xticks(rotation=90)
plt.title('Company vs Laptop Price')
plt.xlabel('Company')
plt.ylabel('Price (Euro)')
plt.show()
```

**What types of laptops are more common in the dataset?**

```python
# Count plot for laptop types
plt.figure(figsize=(12, 6))
sns.countplot(x='TypeName', data=df)
plt.xticks(rotation=90)
plt.title('Distribution of Laptop Types')
plt.xlabel('Laptop Type')
plt.ylabel('Count')
plt.show()
```

**Does screen resolution affect the price?**

```python
# Grouping similar resolutions together for better visualization
df['Resolution_Category'] = df['ScreenResolution'].apply(lambda x: '4K' if '3840' in x or '2160' in x else 'Full HD' if '1920' in x else 'Other')

# Boxplot of Resolution Category vs Price
plt.figure(figsize=(10, 6))
sns.boxplot(x='Resolution_Category', y='Price (Euro)', data=df)
plt.title('Screen Resolution Category vs Laptop Price')
plt.xlabel('Resolution Category')
plt.ylabel('Price (Euro)')
plt.show()
```

**Which operating systems are most common in the dataset?**

**How do different companies compare in terms of weight and price?**

```python
# Scatter plot of Price vs Weight colored by Company
plt.figure(figsize=(10, 6))
for company in df['Company'].unique():
    subset = df[df['Company'] == company]
    plt.scatter(subset['Weight (kg)'], subset['Price (Euro)'], label=company)

plt.title('Price vs Weight by Company')
plt.xlabel('Weight (kg)')
plt.ylabel('Price (Euro)')
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
plt.show()
```
